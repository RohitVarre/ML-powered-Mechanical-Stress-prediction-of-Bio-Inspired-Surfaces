```python
import numpy as np
import pandas as pd
import os
```

```python
left_stress = []
right_stress = []
name = []
data_path=r'C:\Users\Rohit Varre\Desktop\Project Spidy'
categories=os.listdir(data_path)
for folder in categories:
    if folder == 'stress_new':
        folder_path=os.path.join(data_path,folder)
        data=os.listdir(folder_path)
        for dat in data:
            print(dat)
            temp_path = os.path.join(folder_path,dat)
            temp = pd.read_csv(temp_path)
            temp = temp.iloc[7:]
            temp.columns = ['x','y']
            temp = temp.astype(float)
            temp.reset_index(inplace=True)
            left = 0
            right = 0
            for i in range(0,750):
                left = left + temp['y'][i] 
            for i in range(750,1501):
                right = right + temp['y'][i]
            left_stress.append(left)
            right_stress.append(right)
            print(left)
            name.append(dat)
```

```python
detach = pd.DataFrame()
detach['name'] = name
detach['left'] = left_stress
detach['right'] = right_stress
detach['net'] = detach['right'] - detach['left']
print(detach)
```
