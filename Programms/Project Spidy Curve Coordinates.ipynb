```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import LineString
```

```python
def bezier_curve(controls, n = 3000):
    Qx = []
    Qy = []
    ts = []

    Px0 = controls[0]
    Py0 = controls[4]
    Px1 = controls[1]
    Py1 = controls[5]
    Px2 = controls[2]
    Py2 = controls[6]
    Px3 = controls[3]
    Py3 = controls[7]

    cx3=-Px0 + 3*(Px1-Px2) + Px3
    cy3=-Py0 + 3*(Py1-Py2) + Py3
    cx2=3*(Px0-2*Px1+Px2)
    cy2=3*(Py0-2*Py1+Py2)
    cx1=3*(Px1-Px0)
    cy1=3*(Py1-Py0)
    cx0=Px0
    cy0=Py0

    dt = 1/n

    for i in range(0,n+1):
        t=i*dt
        ts.append(t)
        Qx.append(((cx3*t+cx2)*t+cx1)*t + cx0)
        Qy.append(((cy3*t+cy2)*t+cy1)*t + cy0)
    
    return Qx, Qy, ts
```
